```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/Life Expectancy Data.csv")
print(df.columns)
```

```python
df.describe()
```

```python
# list all the missing values in the df

df.isnull().sum()
```

```python
new_column = []
for x in df.columns:
  new_column.append(x.strip())

df.columns = new_column
df
```

```python
df.columns[df.isnull().any()]
```

```python
df['Life expectancy'].fillna(df['Life expectancy'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# list all the columns with null values lesser than 20 and more than 0

null_counts = df.isnull().sum()
cols_with_nulls = null_counts[(null_counts > 0) & (null_counts < 20)]
print(cols_with_nulls)
```

```python
for x in cols_with_nulls.index:
  df[x].fillna(df[x].median(), inplace=True)
```

```python
df['Alcohol'].value_counts()
```

```python
df['Alcohol'].fillna(df['Alcohol'].mode()[0], inplace=True)
```

```python
df['Hepatitis B'].value_counts()
```

```python
Hep_B_mean = (df['Hepatitis B'].value_counts().index[0] + df['Hepatitis B'].value_counts().index[1] + df['Hepatitis B'].value_counts().index[2])/3
```

```python
df['Hepatitis B'].fillna(Hep_B_mean, inplace=True)
```

```python
df['BMI'].value_counts()
```

```python
df['BMI'].fillna(df['BMI'].median(), inplace=True)
```

```python
df['Total expenditure'].fillna(df['Total expenditure'].mean(), inplace=True)
```

```python
df['GDP'].value_counts()
```

```python
# Identify object columns
object_cols = df.select_dtypes(include=['object']).columns

# Create dummy variables for object columns
df = pd.get_dummies(df, columns=object_cols, drop_first=True)

# Now, df contains dummy variables instead of the original object columns
df.head()
```

```python
df = df.replace({False:0, True:1})
df.head()
```
