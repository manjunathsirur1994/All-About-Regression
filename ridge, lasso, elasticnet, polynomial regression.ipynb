```python
import pandas as pd
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PolynomialFeatures
```

```python
df = pd.read_csv(r"/content/Life_expectancy_cleaned_data.csv")
df
```

```python
x = df.drop(['Life expectancy'], axis=1)
y = df['Life expectancy']
```

```python
x[:5]
```

```python
y[:5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = Ridge(alpha=1.0)
model2 = Lasso(alpha=1.0)
model3 = ElasticNet(alpha=1.0, l1_ratio=0.5)
model.fit(x_train, y_train)
model2.fit(x_train, y_train)
model3.fit(x_train, y_train)
```

```python
y_preds = model.predict(x_test)
y_preds2 = model2.predict(x_test)
y_preds3 = model3.predict(x_test)
```

```python
y_preds2[:5]
```

```python
y_preds3[:5]
```

```python
y_test[:5]
```

```python
mean_squared_error(y_test, y_preds)
```

```python
mean_squared_error(y_test, y_preds2)
```

```python
mean_squared_error(y_test, y_preds3)
```

```python
model.score(x_test, y_preds)
```

```python
model2.score(x_test, y_preds2)
```

```python
model3.score(x_test, y_preds3)
```

```python
degree = 2
poly_features = PolynomialFeatures(degree = degree)
x_poly = poly_features.fit_transform(x)
```

```python
x_poly[:5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
y_preds = lr_model.predict(x_test)
y_preds[:5]
```
